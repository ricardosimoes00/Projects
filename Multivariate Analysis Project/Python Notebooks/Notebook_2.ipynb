```python
#@title
import lazypredict
from lazypredict.Supervised import LazyClassifier
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

```python
dataset_train =  pd.read_csv("dataset_train.csv")
dataset_test =  pd.read_csv("dataset_test.csv")
```

```python
dataset_train = dataset_train.drop(labels='Unnamed: 0', axis=1)
dataset_test = dataset_test.drop(labels='Unnamed: 0', axis=1)
```

```python
#X_train, X_test, y_train, y_test = train_test_split(X_standardized, y, test_size=0.2, random_state=1)

X_train = dataset_train.drop(["LET_IS"], axis = 1)
y_train = dataset_train["LET_IS"]
X_test = dataset_test.drop(["LET_IS"], axis = 1)
y_test = dataset_test["LET_IS"]

clf = LazyClassifier()
models, predictions = clf.fit(X_train, X_test, y_train, y_test)
```

```python
models
```

```python
from sklearn.model_selection import cross_val_score, GridSearchCV, StratifiedKFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import roc_curve, precision_recall_curve, auc, make_scorer, recall_score, accuracy_score, precision_score, confusion_matrix,balanced_accuracy_score,f1_score
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import AdaBoostClassifier
import pandas as pd
```

```python
from sklearn.neural_network import MLPClassifier

mlp_g = {
    "hidden_layer_sizes": [(82,),(84,),(86,),(88,),(90,),(92,),(94,)]
}

scorers_mlp = {
    'precision_score': make_scorer(precision_score),
    'recall_score': make_scorer(recall_score),
    'accuracy_score': make_scorer(accuracy_score),
    'balanced_accuracy': make_scorer(balanced_accuracy_score),
    'f1': make_scorer(f1_score)
}

skf = StratifiedKFold(n_splits=5)
refit_score='recall_score'

mlp_gsCV = GridSearchCV(
    MLPClassifier(activation = "tanh", solver = "adam", learning_rate = "adaptive"),
    mlp_g,
    scoring=scorers_mlp,
    refit=refit_score,
    cv = skf
)

mlp_gsCV.fit(X_train, y_train)
```

```python
mlp_results = pd.DataFrame(mlp_gsCV.cv_results_)
mlp_results = pd.DataFrame(mlp_gsCV.cv_results_)[["param_hidden_layer_sizes","mean_test_precision_score", "mean_test_recall_score",'mean_test_f1']]
mlp_results
```

```python
mlp = MLPClassifier(hidden_layer_sizes = (94,), activation = "tanh", solver = "adam", learning_rate = "adaptive", random_state= 123)
mlp_model = mlp.fit(X_train, y_train)
y_pred_mlp  = mlp_model.predict(X_test)


print("MLP:")
print("acc",accuracy_score(y_test, y_pred_mlp))
print("bacc",balanced_accuracy_score(y_test, y_pred_mlp))
print("precision",precision_score(y_test, y_pred_mlp))
print("sensitivity/recall",recall_score(y_test, y_pred_mlp))
print("specificity",balanced_accuracy_score(y_test, y_pred_mlp)*2-recall_score(y_test, y_pred_mlp))
print("fi",f1_score(y_test, y_pred_mlp))
```

```python
from sklearn.tree import DecisionTreeClassifier

dt_g = {"max_depth": [2,3,4,5,6,7,8,9],
        "criterion": ["gini","entropy"]
        
}

scorers_dt = {
    'precision_score': make_scorer(precision_score),
    'recall_score': make_scorer(recall_score),
    'accuracy_score': make_scorer(accuracy_score),
    'balanced_accuracy': make_scorer(balanced_accuracy_score),
    'f1': make_scorer(f1_score)
}

skf = StratifiedKFold(n_splits=5)
refit_score='recall_score'

dt_gsCV = GridSearchCV(
    DecisionTreeClassifier(),
    dt_g,
    scoring=scorers_dt,
    refit=refit_score,
    cv = skf
)

dt_gsCV.fit(X_train, y_train)
```

```python
dt_results = pd.DataFrame(dt_gsCV.cv_results_)
dt_results = pd.DataFrame(dt_gsCV.cv_results_)[["param_criterion","param_max_depth","mean_test_precision_score", "mean_test_recall_score",'mean_test_f1']]
dt_results
```

```python
dt = DecisionTreeClassifier(max_depth=7,criterion="gini",random_state=123)
dt_model = dt.fit(X_train, y_train)
y_pred_dt = dt_model.predict(X_test)

print("DT:")
print("acc",accuracy_score(y_test, y_pred_dt))
print("bacc",balanced_accuracy_score(y_test, y_pred_dt))
print("precision",precision_score(y_test, y_pred_dt))
print("Sensitivity/Recall",recall_score(y_test, y_pred_dt)) # = Sensitivity
print("specificity",balanced_accuracy_score(y_test, y_pred_dt)*2-recall_score(y_test, y_pred_dt))
print("f1",f1_score(y_test, y_pred_dt))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_g = {
    "n_estimators": [2,3,4,5,6],
    "criterion": ["entropy","gini"],
    "max_depth": [15,16,17,18,19,20]

}

scorers_rf = {
    'precision_score': make_scorer(precision_score),
    'recall_score': make_scorer(recall_score),
    'accuracy_score': make_scorer(accuracy_score),
    'balanced_accuracy': make_scorer(balanced_accuracy_score),
    'f1': make_scorer(f1_score)
}

skf = StratifiedKFold(n_splits=5, random_state = 549)
refit_score='recall_score'

rf_gsCV = GridSearchCV(
    RandomForestClassifier(random_state = 549),
    rf_g,
    scoring=scorers_rf,
    refit=refit_score,
    cv = skf
)

rf_gsCV.fit(X_train, y_train)
rf_results = pd.DataFrame(rf_gsCV.cv_results_)
#rf_results = rf_resutlts.sort_values(by='mean_test_precision_score', ascending=False)
```

```python
rf_results = pd.DataFrame(rf_gsCV.cv_results_)[["param_n_estimators","param_max_depth","param_criterion","mean_test_precision_score", "mean_test_recall_score","mean_test_accuracy_score","mean_test_balanced_accuracy", 'mean_test_f1']]
rf_results = rf_results.sort_values(by='mean_test_recall_score', ascending=False)
rf_results
```
