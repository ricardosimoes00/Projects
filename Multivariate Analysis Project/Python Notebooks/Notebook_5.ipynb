```python
import lazypredict
from lazypredict.Supervised import LazyClassifier
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

```python
from sklearn.model_selection import cross_val_score, GridSearchCV, StratifiedKFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import roc_curve, precision_recall_curve, auc, make_scorer, recall_score, accuracy_score, precision_score, confusion_matrix,balanced_accuracy_score,f1_score
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import AdaBoostClassifier
import pandas as pd
```

```python
dataset_train_boruta_NoOUT =  pd.read_csv("dataset_train_famd_NoOUT.csv")
dataset_test_boruta_NoOUT =  pd.read_csv("dataset_test_famd_NoOUT.csv")
```

```python
dataset_train_boruta_NoOUT = dataset_train_boruta_NoOUT.drop(labels='Unnamed: 0', axis=1)
dataset_test_boruta_NoOUT = dataset_test_boruta_NoOUT.drop(labels='Unnamed: 0', axis=1)
dataset_train_boruta_NoOUT = dataset_train_boruta_NoOUT.drop(labels='X', axis=1)
dataset_test_boruta_NoOUT = dataset_test_boruta_NoOUT.drop(labels='X', axis=1)
```

```python
dataset_train_boruta_NoOUT
```

```python
#X = data4

#X=X.drop(["LET_IS"], axis = 1)
#X_standardized = preprocessing.scale(X)
#y = data4["LET_IS"]

#X_train, X_test, y_train, y_test = train_test_split(X_standardized, y, test_size=0.2, random_state=1)

X_train = dataset_train_boruta_NoOUT.drop(["LET_IS"], axis = 1)
y_train = dataset_train_boruta_NoOUT["LET_IS"]
X_test = dataset_test_boruta_NoOUT.drop(["LET_IS"], axis = 1)
y_test = dataset_test_boruta_NoOUT["LET_IS"]

clf = LazyClassifier()
models, predictions = clf.fit(X_train, X_test, y_train, y_test)
```

```python
models
```

```python
for col in X_test:
  X_test[col] = X_test[col].astype(float)
  X_train[col] = X_train[col].astype(float) 
```

```python
mlp_g = {
    "hidden_layer_sizes": [(82,),(84,),(86,),(88,),(90,),(92,),(94,),(96)]
}

scorers_mlp = {
    'precision_score': make_scorer(precision_score),
    'recall_score': make_scorer(recall_score),
    'accuracy_score': make_scorer(accuracy_score),
    'balanced_accuracy': make_scorer(balanced_accuracy_score),
    'f1': make_scorer(f1_score)
}

skf = StratifiedKFold(n_splits=5, random_state = 549)
refit_score='recall_score'

mlp_gsCV = GridSearchCV(
    MLPClassifier(activation = "tanh", solver = "adam", learning_rate = "adaptive", random_state= 123),
    mlp_g,
    scoring=scorers_mlp,
    refit=refit_score,
    cv = skf
)

mlp_gsCV.fit(X_train, y_train)
```

```python
mlp_results = pd.DataFrame(mlp_gsCV.cv_results_)
mlp_results = pd.DataFrame(mlp_gsCV.cv_results_)[["param_hidden_layer_sizes","mean_test_precision_score", "mean_test_recall_score","mean_test_accuracy_score","mean_test_balanced_accuracy", 'mean_test_f1']]
mlp_results
```

```python
mlp = MLPClassifier(hidden_layer_sizes = (84,), activation = "tanh", solver = "adam", learning_rate = "adaptive", random_state= 123)
mlp_model = mlp.fit(X_train, y_train)
y_pred_mlp  = mlp_model.predict(X_test)


print("MLP:")
print("acc",accuracy_score(y_test, y_pred_mlp))
print("bacc",balanced_accuracy_score(y_test, y_pred_mlp))
print("precision",precision_score(y_test, y_pred_mlp))
print("sensitivity/recall",recall_score(y_test, y_pred_mlp))
print("specificity",balanced_accuracy_score(y_test, y_pred_mlp)*2-recall_score(y_test, y_pred_mlp))
print("fi",f1_score(y_test, y_pred_mlp))
```
