```python
%pylab inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import json
import numpy as np
```

```python
#@title
captions = (json.load(open("caps_generated_enc_dec_w_attention.json")))
```

```python
img=mpimg.imread('219.tif')
imgplot = plt.imshow(img)
plt.show()

for e in captions:
  if e["image_name"] == "219.tif":  
    print( [''.join(i) for i in e["caption"]])
```
